```python
%%html
<!DOCTYPE html>

<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Order Statistics Simulator</title>
    <script src="https://cdnjs.cloudflare.com/ajax/libs/Chart.js/3.9.1/chart.min.js"></script>
    <style>
        body{
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            max-width: 1200px;
            margin: 0 auto;
            padding: 20px;
            background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
            min-height: 100vh;
        }
        .container {
            background: white;
            border-radius: 15px;
            padding: 30px;
            box-shadow: 0 10px 40px rgba(0,0,0,0.2);
        }
        h1 {
            color: #333;
            text-align: center;
            margin-bottom: 10px;
        }
        .subtitle {
            text-align: center;
            color: #666;
            margin-bottom: 30px;
            font-size: 14px;
        }
        .controls {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(200px, 1fr));
            gap: 20px;
            margin-bottom: 30px;
            padding: 20px;
            background: #f8f9fa;
            border-radius: 10px;
        }
        .control-group {
            display: flex;
            flex-direction: column;
        }
        label {
            font-weight: 600;
            margin-bottom: 8px;
            color: #333;
            font-size: 14px;
        }
        input[type="number"], input[type="range"] {
            padding: 8px;
            border: 2px solid #ddd;
            border-radius: 5px;
            font-size: 14px;
        }
        input[type="range"] {
            padding: 0;
        }
        .range-value {
            text-align: center;
            font-weight: bold;
            color: #667eea;
            margin-top: 5px;
        }
        button {
            background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
            color: white;
            border: none;
            padding: 12px 30px;
            font-size: 16px;
            border-radius: 8px;
            cursor: pointer;
            font-weight: 600;
            transition: transform 0.2s, box-shadow 0.2s;
            margin: 10px 5px;
        }
        button:hover {
            transform: translateY(-2px);
            box-shadow: 0 5px 15px rgba(102, 126, 234, 0.4);
        }
        button:active {
            transform: translateY(0);
        }
        .chart-container {
            position: relative;
            height: 400px;
            margin-bottom: 30px;
        }
        .stats {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(200px, 1fr));
            gap: 15px;
            margin-top: 20px;
        }
        .stat-box {
            background: linear-gradient(135deg, #667eea15 0%, #764ba215 100%);
            padding: 15px;
            border-radius: 8px;
            border-left: 4px solid #667eea;
        }
        .stat-label {
            font-size: 12px;
            color: #666;
            text-transform: uppercase;
            letter-spacing: 0.5px;
        }
        .stat-value {
            font-size: 24px;
            font-weight: bold;
            color: #333;
            margin-top: 5px;
        }
        .button-group {
            text-align: center;
            margin: 20px 0;
        }
        .info-box {
            background: #e3f2fd;
            padding: 15px;
            border-radius: 8px;
            margin-bottom: 20px;
            border-left: 4px solid #2196f3;
        }
        .formula {
            font-family: 'Courier New', monospace;
            background: #f5f5f5;
            padding: 10px;
            border-radius: 5px;
            margin: 10px 0;
            text-align: center;
        }
    </style>
</head>
<body>
    <div class="container">
        <h1>üìä Order Statistics Beta Distribution Simulator</h1>
        <div class="subtitle">Verify that X‚Çç‚Çñ‚Çé from Uniform(0,1) follows Beta(k, n-k+1)</div>


    <div class="info-box">
        <strong>Theory:</strong> The k-th order statistic X‚Çç‚Çñ‚Çé from n i.i.d. Uniform(0,1) random variables follows a Beta(k, n-k+1) distribution.
        <div class="formula">PDF: f(x) = [n! / ((k-1)!(n-k)!)] √ó x^(k-1) √ó (1-x)^(n-k)</div>
    </div>

    <div class="controls">
        <div class="control-group">
            <label for="n">Sample Size (n):</label>
            <input type="number" id="n" min="2" max="20" value="5">
        </div>
        <div class="control-group">
            <label for="k">Order Statistic (k):</label>
            <input type="number" id="k" min="1" max="5" value="3">
        </div>
        <div class="control-group">
            <label for="simulations">Number of Simulations:</label>
            <input type="range" id="simulations" min="100" max="10000" step="100" value="1000">
            <div class="range-value" id="simValue">1000</div>
        </div>
        <div class="control-group">
            <label for="bins">Histogram Bins:</label>
            <input type="range" id="bins" min="20" max="100" step="5" value="50">
            <div class="range-value" id="binsValue">50</div>
        </div>
    </div>

    <div class="button-group">
        <button onclick="runSimulation()">‚ñ∂ Run Simulation</button>
        <button onclick="clearChart()">üîÑ Clear</button>
    </div>

    <div class="chart-container">
        <canvas id="myChart"></canvas>
    </div>

    <div class="stats">
        <div class="stat-box">
            <div class="stat-label">Simulated Mean</div>
            <div class="stat-value" id="simMean">-</div>
        </div>
        <div class="stat-box">
            <div class="stat-label">Theoretical Mean</div>
            <div class="stat-value" id="theoMean">-</div>
        </div>
        <div class="stat-box">
            <div class="stat-label">Simulated Variance</div>
            <div class="stat-value" id="simVar">-</div>
        </div>
        <div class="stat-box">
            <div class="stat-label">Theoretical Variance</div>
            <div class="stat-value" id="theoVar">-</div>
        </div>
    </div>
</div>

<script>
    let chart = null;

    // Update range value displays
    document.getElementById('simulations').addEventListener('input', function(e) {
        document.getElementById('simValue').textContent = e.target.value;
    });

    document.getElementById('bins').addEventListener('input', function(e) {
        document.getElementById('binsValue').textContent = e.target.value;
    });

    // Update k max value when n changes
    document.getElementById('n').addEventListener('input', function(e) {
        const kInput = document.getElementById('k');
        kInput.max = e.target.value;
        if (parseInt(kInput.value) > parseInt(e.target.value)) {
            kInput.value = e.target.value;
        }
    });

    // Gamma function approximation
    function gamma(z) {
        if (z === 1) return 1;
        if (z === 0.5) return Math.sqrt(Math.PI);
        if (z > 1) return (z - 1) * gamma(z - 1);

        // Stirling's approximation for larger values
        return Math.sqrt(2 * Math.PI / z) * Math.pow((z / Math.E) * Math.sqrt(z * Math.sinh(1/z)), z);
    }

    // Beta distribution PDF
    function betaPDF(x, alpha, beta) {
        if (x <= 0 || x >= 1) return 0;
        const B = (gamma(alpha) * gamma(beta)) / gamma(alpha + beta);
        return Math.pow(x, alpha - 1) * Math.pow(1 - x, beta - 1) / B;
    }

    function runSimulation() {
        const n = parseInt(document.getElementById('n').value);
        const k = parseInt(document.getElementById('k').value);
        const numSims = parseInt(document.getElementById('simulations').value);
        const numBins = parseInt(document.getElementById('bins').value);

        if (k > n) {
            alert('k must be less than or equal to n!');
            return;
        }

        // Run simulations
        const orderStats = [];
        for (let sim = 0; sim < numSims; sim++) {
            // Generate n uniform random variables
            const sample = [];
            for (let i = 0; i < n; i++) {
                sample.push(Math.random());
            }
            // Sort and get k-th order statistic
            sample.sort((a, b) => a - b);
            orderStats.push(sample[k - 1]);
        }

        // Calculate statistics
        const simMean = orderStats.reduce((a, b) => a + b, 0) / numSims;
        const simVar = orderStats.reduce((acc, val) => acc + Math.pow(val - simMean, 2), 0) / numSims;

        // Theoretical values for Beta(k, n-k+1)
        const alpha = k;
        const beta = n - k + 1;
        const theoMean = alpha / (alpha + beta);
        const theoVar = (alpha * beta) / (Math.pow(alpha + beta, 2) * (alpha + beta + 1));

        // Update stats display
        document.getElementById('simMean').textContent = simMean.toFixed(4);
        document.getElementById('theoMean').textContent = theoMean.toFixed(4);
        document.getElementById('simVar').textContent = simVar.toFixed(4);
        document.getElementById('theoVar').textContent = theoVar.toFixed(4);

        // Create histogram
        const binWidth = 1 / numBins;
        const histogram = new Array(numBins).fill(0);

        orderStats.forEach(val => {
            const binIndex = Math.min(Math.floor(val / binWidth), numBins - 1);
            histogram[binIndex]++;
        });

        // Normalize histogram to probability density
        const histogramDensity = histogram.map(count => count / (numSims * binWidth));

        // Generate theoretical Beta PDF
        const theoreticalX = [];
        const theoreticalY = [];
        for (let i = 0; i <= 100; i++) {
            const x = i / 100;
            theoreticalX.push(x);
            theoreticalY.push(betaPDF(x, alpha, beta));
        }

        // Create chart
        const ctx = document.getElementById('myChart').getContext('2d');

        if (chart) {
            chart.destroy();
        }

        const binLabels = [];
        for (let i = 0; i < numBins; i++) {
            binLabels.push((i * binWidth + binWidth / 2).toFixed(3));
        }

        chart = new Chart(ctx, {
            type: 'bar',
            data: {
                labels: binLabels,
                datasets: [{
                    label: 'Simulated Distribution',
                    data: histogramDensity,
                    backgroundColor: 'rgba(102, 126, 234, 0.6)',
                    borderColor: 'rgba(102, 126, 234, 1)',
                    borderWidth: 1,
                    barPercentage: 1,
                    categoryPercentage: 1
                }, {
                    label: 'Theoretical Beta PDF',
                    data: theoreticalX.map((x, i) => ({x: x, y: theoreticalY[i]})),
                    type: 'line',
                    borderColor: 'rgb(255, 99, 132)',
                    backgroundColor: 'rgba(255, 99, 132, 0.1)',
                    borderWidth: 3,
                    pointRadius: 0,
                    fill: false,
                    tension: 0.4
                }]
            },
            options: {
                responsive: true,
                maintainAspectRatio: false,
                plugins: {
                    title: {
                        display: true,
                        text: `Distribution of X(${k}) from ${numSims} simulations (n=${n}) vs Beta(${alpha}, ${beta})`,
                        font: { size: 16, weight: 'bold' }
                    },
                    legend: {
                        display: true,
                        position: 'top'
                    }
                },
                scales: {
                    x: {
                        title: {
                            display: true,
                            text: 'Value',
                            font: { size: 14, weight: 'bold' }
                        },
                        ticks: {
                            maxTicksLimit: 10
                        }
                    },
                    y: {
                        title: {
                            display: true,
                            text: 'Probability Density',
                            font: { size: 14, weight: 'bold' }
                        },
                        beginAtZero: true
                    }
                }
            }
        });
    }

    function clearChart() {
        if (chart) {
            chart.destroy();
            chart = null;
        }
        document.getElementById('simMean').textContent = '-';
        document.getElementById('theoMean').textContent = '-';
        document.getElementById('simVar').textContent = '-';
        document.getElementById('theoVar').textContent = '-';
    }

    // Run initial simulation
    runSimulation();
</script>


</body>
</html>
```
